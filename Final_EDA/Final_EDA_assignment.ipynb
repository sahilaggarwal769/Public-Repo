```python
# Referred Kaggle and iris dataset example.
import pandas as pd
# we can read the data directly from raw github link
# we are also defining the name of the columns.
#make sure that your csv file and ipynb notebook are in the same folder. If they are in different folder then you have to define the complete path
df=pd.read_csv('haberman.csv',names=["age","operation_Year","axil_nodes","survival_status"],header=0)
df.head()
```

```python
print("Shape of dataset is: ",df.shape)
print()
print("Columns are: ",df.columns)
print()
print("Info about the dataset: ",df.info())
# No null values are present in dataset
print()
print("Value Count: ",df['survival_status'].value_counts())
print()
print("People belonging to class 1(people who survived more than 5 yrs of operation) are: ",df['survival_status'].value_counts()[1])
print("People belonging to class 2(people who survived less than 5 yrs of operation) are: ",df['survival_status'].value_counts()[2])

# The value count for 1 (survived more than 5 yrs) are 225 and person who survived less than 5 yrs are 81
#Binary classification
print()
print("Percentage of people who survived more than 5 yrs of operation are:",225/(225+81)*100)
print("Percentage of people who survived less than 5 yrs of operation are:",81/(225+81)*100)
# Dataset is unbalanced
```

## Observations :
1. The value count for 1 (survived more than 5 yrs) are 225 and person who survived less than 5 yrs are 81.

To predict whether the person will survive more than 5 years after the operation based on person's age, year of operation and axil nodes

```python
# Plotting Histograms for different variables
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set_style("whitegrid")
sns.FacetGrid(df, hue="survival_status",size=4).map(sns.distplot,"age").add_legend()
plt.title("Histogram of Age Of Person")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df, hue="survival_status",size=4).map(sns.distplot,"operation_Year").add_legend()
plt.title("Histogram of Operation Year")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df, hue="survival_status",size=4).map(sns.distplot,"axil_nodes").add_legend()
plt.title("Histogram of Axil nodes")
plt.show()
# Axil node = 0 has highest survival rate
```

## Observations:
1. Axil node = 0 has highest survival rate
2. No other histogram provide much information.

```python
# Calculating pdf's
# plotting for class-1
counts, bin_edges = np.histogram(class_1['age'], bins=10, 
                                 density = True)
#print(counts)
pdf=counts/sum(counts)
#print(pdf)
#print(bin_edges)
cdf=np.cumsum(pdf)
#print(np.cumsum(pdf))

plt.plot(bin_edges[1:],cdf)
#plt.plot(bin_edges[1:],cdf)

# plotting for class-2
counts, bin_edges = np.histogram(class_2['age'], bins=10, 
                                 density = True)
#print(counts)
pdf=counts/sum(counts)
#print(pdf)
#print(bin_edges)
cdf=np.cumsum(pdf)
#print(np.cumsum(pdf))

plt.plot(bin_edges[1:],cdf)
#plt.plot(bin_edges[1:],cdf)

plt.title("CDF of age of person")
plt.xlabel("Age")
plt.legend(['Survived', 'Not-Survived'])

plt.show()
```

```python
# plotting for class-1
counts, bin_edges = np.histogram(class_1['operation_Year'], bins=10, 
                                 density = True)
#print(counts)
pdf=counts/sum(counts)
#print(pdf)
#print(bin_edges)
cdf=np.cumsum(pdf)
#print(np.cumsum(pdf))



plt.plot(bin_edges[1:],cdf)
#plt.plot(bin_edges[1:],cdf)

# plotting for class-2
counts, bin_edges = np.histogram(class_2['operation_Year'], bins=10, 
                                 density = True)
#print(counts)
pdf=counts/sum(counts)
#print(pdf)
#print(bin_edges)
cdf=np.cumsum(pdf)
#print(np.cumsum(pdf))

plt.plot(bin_edges[1:],cdf)
#plt.plot(bin_edges[1:],cdf)

plt.title("CDF of Operation Year of person")
plt.xlabel("Operation Year")
plt.legend(['Survived', 'Not-Survived'])
plt.show()
```

```python
# plotting for class-1
counts, bin_edges = np.histogram(class_1['axil_nodes'], bins=10, 
                                 density = True)
#print(counts)
pdf=counts/sum(counts)
#print(pdf)
#print(bin_edges)
cdf=np.cumsum(pdf)
#print(np.cumsum(pdf))



plt.plot(bin_edges[1:],cdf)
#plt.plot(bin_edges[1:],cdf)

# plotting for class-2
counts, bin_edges = np.histogram(class_2['axil_nodes'], bins=10, 
                                 density = True)
#print(counts)
pdf=counts/sum(counts)
#print(pdf)
#print(bin_edges)
cdf=np.cumsum(pdf)
#print(np.cumsum(pdf))

plt.plot(bin_edges[1:],cdf)
#plt.plot(bin_edges[1:],cdf)

plt.title("CDF of Axil nodes of person")
plt.xlabel("Axil nodes")
plt.legend(['Survived', 'Not-Survived'])
plt.show()
# Apprx 90% person with axil nodes less than 10 survived.
```

## Observations:
1. Plotting cdf among age shows that person with age less than 34 were able to survive while person with age more than 77 were not able to survive.
2. Plotting cdf among axil nodes shows that person with axil nodes less than 10 are having 90% chance of survival.

```python
# Visualization via Box-plot
sns.boxplot(x="survival_status",y="age",data=df)
plt.title("Box-Plot for survival status and age")
plt.show()
sns.boxplot(x="survival_status",y="operation_Year",data=df)
plt.title("Box-Plot for survival status and Operation Year")
plt.show()
sns.boxplot(x="survival_status",y="axil_nodes",data=df)
plt.title("Box-Plot for survival status and Axil nodes")
plt.show()
#Person with age greater than 77 were not able to survive
#As axil node increases chances of death increases
```

## Observations:
1. Persons with age b/w 30 and 35 survived while people with age greater than 77 were not able to survive.
2. As the axil node increases chances of survival decreases.
3. There are a lot of outliers in axil node features for the people who survived more than 5 yrs.
